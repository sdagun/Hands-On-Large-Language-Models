```python
# Python code to demonstrate word embeddings with visualization
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import normalize
import numpy as np

# Dummy word embeddings (vectors in 5D space)
embeddings = {
    "king": np.array([0.8, 0.2, 0.1, 0.5, 0.9]),
    "queen": np.array([0.9, 0.3, 0.2, 0.6, 1.0]),
    "man": np.array([0.7, 0.1, 0.0, 0.4, 0.8]),
    "woman": np.array([0.85, 0.25, 0.15, 0.55, 0.95]),
    "apple": np.array([0.1, 0.8, 0.7, 0.9, 0.1]),
    "orange": np.array([0.2, 0.85, 0.75, 0.95, 0.15]),
}

# Normalize and reduce to 2D for plotting
words = list(embeddings.keys())
X = np.array([embeddings[word] for word in words])
X_norm = normalize(X)
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_norm)

# Plotting
plt.figure(figsize=(8, 6))
for i, word in enumerate(words):
    x, y = X_reduced[i]
    plt.scatter(x, y, label=word)
    plt.text(x + 0.01, y + 0.01, word, fontsize=12)

plt.title("2D Visualization of Sample Word Embeddings")
plt.xlabel("PCA Dimension 1")
plt.ylabel("PCA Dimension 2")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
